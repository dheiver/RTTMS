```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Define the function to calculate true stress
def calcular_tensao_verdadeira(tensao, coef_erro, grad_temp, grad_pressao, desvio_ruido, forca, kt, kp, kf, dia_interno, dia_externo):
    # Simulate measurement error
    erro = np.random.normal(0, coef_erro * tensao)

    # Calculate components
    temp_comp = kt * grad_temp * tensao + 0.001 * grad_temp**2 * tensao**2
    pressao_comp = kp * grad_pressao * tensao + 0.0005 * grad_pressao**2 * tensao**2
    forca_comp = kf * forca

    # Calculate the area of the riser
    area = np.pi * 0.25 * (dia_externo**2 - dia_interno**2)

    # Calculate stress due to force
    tensao_forca = forca / area

    # Calculate true stress by combining all components
    tensao_verdadeira = tensao - erro + 0.005 * tensao**2 + temp_comp + pressao_comp + forca_comp + tensao_forca + np.random.normal(0, desvio_ruido)

    return tensao_verdadeira


# Define the function to create gradients
def criar_gradientes(num_segmentos):
    # Generate random amplitudes for temperature and pressure gradients
    temp_amplitude = 0.02 * np.random.uniform(0.9, 1.1, num_segmentos)
    pressao_amplitude = 0.002 * np.random.uniform(0.9, 1.1, num_segmentos)
    
    # Set frequencies and offsets for gradients
    temp_frequencia = 0.1
    pressao_frequencia = 0.02
    temp_offset = 0.01
    pressao_offset = 0.001
    
    # Create time arrays for gradients
    temp_tempo = np.linspace(0, num_segmentos, num_segmentos)
    pressao_tempo = np.linspace(0, num_segmentos, num_segmentos)
    
    # Calculate temperature and pressure gradients
    grad_temp = temp_amplitude * np.sin(2 * np.pi * temp_frequencia * temp_tempo) + temp_offset
    grad_pressao = pressao_amplitude * np.sin(2 * np.pi * pressao_frequencia * pressao_tempo) + pressao_offset
    
    return grad_temp, grad_pressao

# Define the function to create the riser
def criar_riser(params, num_segmentos, tensao_medida, coef_erro, desvio_ruido, forca, dia_interno, dia_externo):
    kt, kp, kf, dia_interno, dia_externo = params

    # Generate temperature and pressure gradients
    grad_temp_cabeca, grad_pressao_cabeca = criar_gradientes(num_segmentos)
    
    # Initialize an array to store true stress values
    tensao_cabeca = np.zeros(num_segmentos)
    
    # Calculate true stress for each segment
    for i in range(num_segmentos):
        tensao_cabeca[i] = calcular_tensao_verdadeira(tensao_medida, coef_erro, grad_temp_cabeca[i], grad_pressao_cabeca[i], desvio_ruido, forca, kt, kp, kf, dia_interno, dia_externo)
    
    return tensao_cabeca

# Define the function to optimize parameters
def otimizar_parametros():
    # Parameters
    NUM_SEGMENTOS = 10
    TENSAO_MEDIDA = 105.0
    COEF_ERRO = 0.02
    DESVIO_RUIDO = 1.0
    FORCA = 500.0
    DIA_INTERNO_CABECA = 0.35
    DIA_EXTERNO_CABECA = 0.45

    # Initial guesses for optimization
    initial_guess = [0.00022, 0.000055, 0.0012, 0.34, 0.44]

    # Bounds for the parameters
    bounds = [(0.0001, 0.0003), (0.00001, 0.0001), (0.0005, 0.0015), (0.3, 0.4), (0.4, 0.5)]

    # Optimize the parameters to minimize the objective function
    def objective_function(params):
        tensao_cabeca = criar_riser(params, NUM_SEGMENTOS, TENSAO_MEDIDA, COEF_ERRO, DESVIO_RUIDO, FORCA, DIA_INTERNO_CABECA, DIA_EXTERNO_CABECA)
        return -np.max(tensao_cabeca)

    result = minimize(objective_function, initial_guess, bounds=bounds)

    return result.x

def criar_visualizacoes_tempo(tensao_cabeca, grad_temp_cabeca, grad_pressao_cabeca, dias):
    # Create subplots
    fig, axs = plt.subplots(2, 2, figsize=(12, 8))

    # Plot true stress over time
    axs[0, 0].plot(np.arange(dias), tensao_cabeca)
    axs[0, 0].set_title('True Stress Over Time')
    axs[0, 0].set_xlabel('Days')
    axs[0, 0].set_ylabel('True Stress')

    # Plot temperature gradient over time
    axs[0, 1].plot(np.arange(dias), grad_temp_cabeca)
    axs[0, 1].set_title('Temperature Gradient Over Time')
    axs[0, 1].set_xlabel('Days')
    axs[0, 1].set_ylabel('Temperature Gradient')

    # Plot pressure gradient over time
    axs[1, 0].plot(np.arange(dias), grad_pressao_cabeca)
    axs[1, 0].set_title('Pressure Gradient Over Time')
    axs[1, 0].set_xlabel('Days')
    axs[1, 0].set_ylabel('Pressure Gradient')

    # Plot combined temperature and pressure gradients
    axs[1, 1].plot(np.arange(dias), grad_temp_cabeca, label='Temperature Gradient', linestyle='--')
    axs[1, 1].plot(np.arange(dias), grad_pressao_cabeca, label='Pressure Gradient', linestyle='-.')
    axs[1, 1].set_title('Combined Temperature and Pressure Gradients Over Time')
    axs[1, 1].set_xlabel('Days')
    axs[1, 1].set_ylabel('Gradient')
    axs[1, 1].legend()

    # Add spacing between subplots
    plt.tight_layout()

    # Save the plot to a file (optional)
    plt.savefig('riser_visualizations.png')

    # Show the plot
    plt.show()

# Find the optimal parameters
optimal_params = otimizar_parametros()

# Create and plot data with the optimal parameters
tensao_cabeca = criar_riser(optimal_params, NUM_SEGMENTOS, TENSAO_MEDIDA, COEF_ERRO, DESVIO_RUIDO, FORCA, DIA_INTERNO_CABECA, DIA_EXTERNO_CABECA)

# Create visualizations with the x-axis representing time in days
dias = 10  # You can adjust the number of days as needed
criar_visualizacoes_tempo(tensao_cabeca, grad_temp_cabeca, grad_pressao_cabeca, dias)
```

```python
def plot_histogram(tensao_cabeca):
    plt.figure(figsize=(8, 6))
    plt.hist(tensao_cabeca, bins=20, color='blue', alpha=0.7)
    plt.title('Histogram of True Stress')
    plt.xlabel('True Stress')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

```python
plot_histogram(tensao_cabeca)
```

```python
def plot_psd(grad_temp, grad_pressao):
    freq_temp, psd_temp = plt.psd(grad_temp, NFFT=len(grad_temp), Fs=1.0)
    freq_pressao, psd_pressao = plt.psd(grad_pressao, NFFT=len(grad_pressao), Fs=1.0)
    
    plt.figure(figsize=(10, 6))
    plt.semilogy(freq_temp, psd_temp, label='Temperature Gradient', color='blue')
    plt.semilogy(freq_pressao, psd_pressao, label='Pressure Gradient', color='red')
    plt.title('Power Spectral Density of Gradients')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('PSD (dB/Hz)')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
plot_psd(grad_temp_cabeca, grad_pressao_cabeca)
```
